###### Wrangling process:
- extract the car model from CarName.
- encode fueltype, aspiration, doornumber and enginelocation in simple [0,1] values.
- encode carbody into four categories, we'll combine the least two values as 'other'. [0,1,2,3]
- encode drivewheel [0,1,2]
- encode enginetype, cylindernumber and fuelsystem into three categories with the last one combining the least values as 'other' [0,1,2,3]

```python
cat_df['car_model'] = cat_df['CarName'].str.split().str.get(0)
```
